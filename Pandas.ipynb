```python
#Create an empty Series
s =pd.Series()
print(s)
```

```python
# creating a seires from ndarray
#if data is an ndarray, then index passed must be of the same lenght. If no index is passed, then by default index will be range(n) where nn is array lenght, i.e., [0,1,2,3...... range(len(array))-1]
import pandas as pd
import numpy as np
data1 = ['a','b','c','d', 'e']
data=np.array(['a','b','c','d','e','f'])
print(data,data.shape)
print(data1)

s=pd.Series(data)
print(s)
```

```python
#customize numbered index in Series
data=np.array(['a','b','c','d'])
s =pd.Series(data, index =[9,10,11,12])
print(s)
print(s[10])
```

```python
#creating a Dictionary in Series
data = {'a':0., 'b': 1., 'c':2., 'd':3.}
s=pd.Series(data)
print(s)

Aa=np.array(data)
print(Aa)
```

```python
#custumizing index in Series
data=np.array(['a','b','c','d'])
s =pd.Series(data, index =['d','c','b','a'])
print(s)
print(s['a'])
```

```python
#Creating a series directly
s=pd.Series([11,12,31], index=['a','b','c'])
print(s)
```

```python
import pandas as pd
import numpy as np
example = pd.DataFrame({
  "name": ["Batman", "Superman", "Joker"],
    "toy" : ["Batmobile", np.nan, "HAHA"],
    "born" : [pd.Timestamp("1956-06-26"),pd.NaT,  pd.NaT]
})
```

```python
example
```

```python
#Drop the rows where at least one element is missing:
example.dropna()
example
```

```python
#droping the columns where at least one element is missing:
example.dropna(axis = "columns")
```

```python
#droping the row where every element is missing:
example.dropna(how = "all")
```

```python
#kdroping rows with 2 nan values:
example.dropna(thresh=2)
```

```python
example.dropna(thresh =3)
```

```python
#Define in which columns to look for missing values:
example.dropna(subset= ['name', 'born'])
```

```python
#Keep the DataFrame with valid Entries in the same variable:
example.dropna(inplace=True)
example
```
